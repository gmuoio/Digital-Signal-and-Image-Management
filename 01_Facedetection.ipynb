## Face detector

```python
detector = MTCNN() # Multi-Task Cascaded Convolutional Neural Network
```

```python
# Funzione per il ritagli dei volti dalle foto

def face_detector(read_path, write_path, names):
    fotosalvate = 0
    fotoscartate= 0
    for name in names:
        i = 0
        for f in sorted(os.listdir("/content/"+ read_path + name + '/')):
            #if f.endswith('.JPG') or f.endswith('.PNG') or f.endswith('.jpg'):
                img = cv.imread("/content/"+ read_path + name + '/' + f)
                face = detector.detect_faces(img)

                if not face or face[0]==False:
                    fotoscartate+=1
                    continue

                #  bounding box extraction
                x1, y1, width, height = face[0]['box']
                x2, y2 = x1 + width, y1 + height

                if width < 50 or height < 50:
                    fotoscartate+=1
                    continue

                # face extraction
                crop = img[y1:y2, x1:x2]
                # plt.imshow(crop); plt.show()

                # Saving cropped images
                # conversion in array of the image
                # face_array = asarray(crop)    
        
                # Saving of the pictures extracted
                os.chdir("/content/" + write_path)
                cv.imwrite(name + "_"+  str(i) + '.jpg', crop)

                fotosalvate+=1
                i+=1

    return fotosalvate,fotoscartate
```

```python
read_path = "drive/MyDrive/Progetto DIGITAL/02_Classification - 2D/Foto/"
write_path="drive/MyDrive/Progetto DIGITAL/02_Classification - 2D/Faces/"
```

```python
#names = ['Anna','Beatrice','Guglielmo']
names = ['Altro']
fotosalvate,fotoscartate = face_detector(read_path, write_path, names)
```
