## Caricamento faces vips

```python
def identity(image):
    return kimage.img_to_array(image).flatten()

# Data loader
def load_data(feature_extractor=identity):
    path = '/content/drive/MyDrive/Progetto DIGITAL/03_Retrieval/vip_faces_20/'

    # Results
    Vip_features = pd.DataFrame(columns=['Features','Label'])


    for f in sorted(os.listdir(path)):
  
        if f[0].isdigit():

            # Carica file ed estraine le features
            image = kimage.load_img(path + f, target_size=(224, 224))
            cur_features = feature_extractor(image)

            Vip_features = Vip_features.append({'Features': cur_features, 'Label': f}, ignore_index=True)

                
    #features = np.array(features)
    return Vip_features
```

```python
def VGGFace_features(img):
    x = kimage.img_to_array(img)
    x = utils.preprocess_input(x)
    x = np.expand_dims(x, axis=0)
    f = vgg_features.predict(x)
    return f.flatten()
```
